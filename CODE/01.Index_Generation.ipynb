# Function

```python
def entropy_weight(data):
    data = (data-data.min())/(data.max()-data.min())
    m,n = data.shape
    data1 = data.values
    k = 1/np.log(m)
    yij = data1.sum(axis = 0)
    
    pij = data/yij
    test = pij*np.log(pij)
    test = np.nan_to_num(test)
    
    ej = -k*(test.sum(axis = 0))
    wi = (1-ej)/np.sum(1-ej)
    wi_list = list(wi)
    return wi_list
```
