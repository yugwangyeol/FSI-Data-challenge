# Data Merging

### Grandata1 & 2 Merge

```python
# grandata 1 시군구 단위로 지표 및 병합 코드 만들기

gran1_df = pd.merge(grandata_df1 , matching_df, on = 'HOME_BIZ_CELL', how = 'left')
gran1_df = gran1_df[gran1_df['BS_YR_Q'] == '204Q'].reset_index(drop = True)

tp_cat = gran1_df.groupby('HOME_BIZ_CELL')[['행정동코드' , '시군구명']].first().reset_index()\
.drop('HOME_BIZ_CELL', axis = 1)
```

```python
gran1_df = gran1_df.groupby('HOME_BIZ_CELL').sum().drop(['LS', '행정동코드'], axis = 1).reset_index()
gran1_df = pd.concat([tp_cat, gran1_df], axis = 1)
```

```python
# grandata 2와 1을 합치고 행정동명까지 groupby 하기
main_df = pd.merge(gran1_df , grandata_df2, on = 'HOME_BIZ_CELL', how = 'left')
main_df = main_df.groupby([ '시군구명']).mean().reset_index()

# 시군구명이 0인 것은 세종특별시를 의미
main_df.loc[main_df.시군구명 == 0, '시군구명'] = '세종특별시'
```

메인 grandata와 한투정보를 합치려면

한투 입장에서 
- 부천시 소사구 부천시 오정구 부천시 원미구 신안군이 없음

grandata 입장에서
- 미추홀구 부천시 봉화군 울릉군 부재

붙이고 null값 처리 요망

```python
hantoo_df.loc[hantoo_df['시군구'] == '부천시 소사구', '시군구'] = '부천시'
hantoo_df.loc[hantoo_df['시군구'] == '부천시 오정구', '시군구'] = '부천시'
hantoo_df.loc[hantoo_df['시군구'] == '부천시 원미구', '시군구'] = '부천시'

# 부천시의 구들은 부천시로 통일
tp_hantoo = hantoo_df.groupby(['시군구'])[['국내주식', '해외주식','연금' ,'채권', '금융거래 다양성', '금융 거래 금액']]\
.mean().reset_index()
```

```python
main_df = pd.merge(main_df ,tp_hantoo, left_on='시군구명', right_on='시군구', how = 'left').drop('시군구', axis = 1)
```

```python
## knn imputation
def imputation(df):
    sigungoo = main_df.iloc[:,:2]    
    df = main_df.iloc[:,2:]
    
    scaler = MinMaxScaler()
    scaled_df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)

    imputer = KNNImputer(n_neighbors=5)
    
    imputed_df = pd.DataFrame(imputer.fit_transform(scaled_df), columns = scaled_df.columns)
    imputed_df = pd.DataFrame(scaler.inverse_transform(imputed_df), columns = imputed_df.columns)
    imputed_df = pd.concat([sigungoo, imputed_df], axis = 1)
    return imputed_df
```

```python
main_df = imputation(main_df)
```

### KB 손해보험 Merge

```python
tp_kookmin_insurance = kookmin_insurance.groupby(['시군구'])[['보험매출', '보험건수']].mean().reset_index()
```

```python
main_df = pd.merge(main_df ,tp_kookmin_insurance, left_on='시군구명', right_on='시군구', how = 'left').drop('시군구', axis = 1)
```

### 국민카드 가맹점  Merge

```python
tp_kb_store = kookmin_store_df[(kookmin_store_df.기준년월 == 202010)|(kookmin_store_df.기준년월 == 202011)|(kookmin_store_df.기준년월 == 202012)]
tp_kb_store.reset_index(drop=True, inplace = True)

kb_card_pivot = pd.pivot_table(data = tp_kb_store, 
               columns = '업종대분류', index = '시군구', values = '총매출금액', aggfunc='mean')[['교통/숙박/여행',
                                                                                   '여가/오락', '의료', '의류/패션잡화']].reset_index()
```

```python
# KNNImputer로 결측치 처리

kb_sigungoo = kb_card_pivot.iloc[:, :1]
kb_card_pivot = kb_card_pivot.iloc[:, 1:]
scaler = MinMaxScaler()
kb_card_pivot = pd.DataFrame(scaler.fit_transform(kb_card_pivot), columns = kb_card_pivot.columns)

imputer = KNNImputer(n_neighbors=5)
kb_card_pivot_imputed = pd.DataFrame(imputer.fit_transform(kb_card_pivot), columns = kb_card_pivot.columns)
kb_card_pivot_imputed = pd.DataFrame(scaler.inverse_transform(kb_card_pivot_imputed), columns = kb_card_pivot_imputed.columns)
kb_card_pivot_imputed = pd.concat([kb_sigungoo,kb_card_pivot_imputed], axis = 1)
```

```python
main_df = pd.merge(main_df ,kb_card_pivot_imputed, left_on='시군구명', right_on='시군구', how = 'left').drop('시군구', axis = 1)
main_df = imputation(main_df)
```
